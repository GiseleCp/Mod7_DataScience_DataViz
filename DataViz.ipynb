```python
import pandas as pd
```

# Importação e preparação dos dados

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
# analisar as tendencias de migração do Brasil para Canadá ao longo dos anos
df.set_index('País', inplace=True) # seta o indice para coluna pais / inplace seta direto no dataframe sem necessidade de criar um novo
```

```python
# intervalo de anos
anos = list(map(str, range(1980, 2014)))
```

```python
# seleção dos dados do brasil com intervalo de datas
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
# criando um dicionario, transformando a series em dataframe
brasil_dict = {'ano' : brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()} #tolist faz conversao da series em lista
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

# Gráficos

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4)) # valores em polegadas, (largura, altura)
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Migração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show() # qualquer info abaixo do show, não aparece no gráfico
```

```python
#plota uma figura, ax é o eixo
#ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker = 'o') # lw - espessura da linha | marker - marcador
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker = 'o', color='g')
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left') # \n - quebra linha
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # multiplos do eixo x é 5
ax.yaxis.set_major_locator(plt.MultipleLocator(500)) # multiplos do eixo y é 500
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
#plt.grid(linestyle='--')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('migracao_brasil_canada_1.png', transparent=False, dpi=300, bbox_inches='tight') # salvar figura

plt.show()
```

```python
# boxplot mostra estatistica com relação aos dados
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].yaxis.set_major_locator(plt.MultipleLocator(500))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Bloxplot da Imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

fig.savefig('migracao_brasil_canada_2.png', transparent=False, dpi=300, bbox_inches='tight') # salvar figura

plt.show()
```

# Gráficos - Comparativo dos países América do Sul

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
# gráfico contendo quantidades, ideal usar gráfico de barras

listaColors = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))
ax.bar(america_sul.index, america_sul['Total'], color = listaColors)
ax.set_title('Imigração da América do Sul para o Canadá de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('País', fontsize=10)
ax.set_ylabel('Número de Imigrantes', fontsize=10)

ax.xaxis.set_tick_params(rotation=45)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

ax.grid(axis='y')
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values(by='Total', ascending=True)
```

```python
# gráfico contendo quantidades, ideal usar gráfico de barras

listaColors = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    listaColors.append('royalblue')
  else:
    listaColors.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = listaColors)
ax.set_title('América do Sul: Brasil é o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=10)
ax.set_ylabel('')

ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=8, ha='left', va='center', fontweight='bold') # v+20 indica onde colocar o texto,

ax.set_frame_on(False) #elina caixa box externo do grafico
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('migracao_brasil_canada_3.png', transparent=False, dpi=300, bbox_inches='tight') # salvar figura

plt.show()
```

# Gráfico animado

```python
import plotly.graph_objs as go
```

```python
dados_brasil['ano'] = dados_brasil['ano'].astype(int)
```

```python
#  figura vazia é criada
fig = go.Figure()
```

```python
#  uma linha é adicionada ao gráfico usando a função fig.add_trace().
#Nesta função, é passado um objeto Scatter, que recebe como argumentos os dados para os eixos X e Y do gráfico.

fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)


# ajuste de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=18)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),


# É adicionado um botão "Play" para a animação usando o argumento updatemenus
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],

    width=1000,
    height=500
)


# configurações de animação são definidas
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames


# mostrando a figura
fig.show()
```
